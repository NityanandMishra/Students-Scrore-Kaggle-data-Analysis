# drop unnamed column

```python
df=df.drop("Unnamed: 0", axis=1)
df
```

```python
#change weekly study hours columns
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

```python
#gender distribution
```

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show
plt.title("Gender DIstribution")
```

```python
#from the above chart we have analysed the number of charts we have analysed that:
#the number of females in the datta is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb
```

```python
sns.heatmap(gb)
plt.show()
```

```python
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
plt.figure(figsize = (4,4))
```

```python
# from the above chart we have concluded that the parents education plays an import role in the students scrore.
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1
```

```python
sns.heatmap(gb1)
plt.show()
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
plt.figure(figsize = (3,3))
```

```python
#Since there's very less variation in the marks of the students we can say that the parents marital status has very less impact
#on the students score.
```

```python
gb2 = df.groupby("NrSiblings").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb2
```

```python
sns.heatmap(gb2, annot=True)
plt.title("Relationship between Number of Siblings and Students Scores")
plt.figure(figsize=(3,3))
plt.show()
```

```python
#from the above we can see that the students of parents having 7 siblings yet can get 68 marks on the other there are students
#with 1 siblings getting 66 marks hence we can concluded the difference is nenegligible and hence number of siblings does not
#impacts the students marks
```

# Distribution of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l= ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.pie(mlist, labels =l, autopct ="%1.2f%%")
plt.title("Enthnic Group Distribution")
plt.show()
```
